```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
actual=pd.read_excel('Downloads/mortalidad_2019.xlsx',sheet_name='morta')
```

```python
actual
```

```python
actual.drop(['Código (CIE10)','Unnamed: 0'],axis=1,inplace=True)
```

```python
actual
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x='No. Casos',y='CAUSA',data=actual)
plt.savefig('Causa_casos.png', bbox_inches="tight")
```

```python
plt.figure(figsize=(15,8))
ax = sns.barplot(x='Tasas %',y='CAUSA',data=actual)
plt.savefig('Tasas_causa.png', bbox_inches="tight")
```
