```python
t2=tasamort[tasamort['Entidad de Residencia'] != 'Nacional']
```

```python
t3=t2.copy()
t3['Entidad de Residencia']=t3['Entidad de Residencia'].replace('México', 'Estado de México')
t3['Entidad de Residencia']=t3['Entidad de Residencia'].replace('Distrito Federal', 'CDMX')
t3.head()
```

```python
t3['Entidad de Residencia'].unique()
```

```python
año_inicio=2000

df=(t3[t3.Periodo.eq(año_inicio)].sort_values(by='Defunciones por Enfermedades Cardiovasculares', ascending=True))

fig, ax=plt.subplots(figsize=(20,15))
ax.barh(df['Entidad de Residencia'], df['Defunciones por Enfermedades Cardiovasculares']);
```

```python
colors=dict(zip(['CDMX', 'Estado de México', 'Jalisco', 'Veracruz Llave', 'Nuevo León', 'Puebla', 'Guanajuato','Michoacán de Ocampo','Chihuahua','Oaxaca'],
                ['#adb0ff', '#ffc7f1', '#90d595', '#e48381', '#aafbff', '#85e1cf', '#eafb50','#33a4ff','#1c9723','#931996']))
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

def draw_barchart(año):
    dff = t3[t3['Periodo'].eq(año)].sort_values(by='Defunciones por Enfermedades Cardiovasculares', ascending=True).tail(10) #Tail como head
    ax.clear()
    ax.barh(dff['Entidad de Residencia'], dff['Defunciones por Enfermedades Cardiovasculares'],color=[colors[x] for x in dff['Entidad de Residencia']])
    dx = dff['Defunciones por Enfermedades Cardiovasculares'].max() / 200
    for i, (value, name) in enumerate(zip(dff['Defunciones por Enfermedades Cardiovasculares'], dff['Entidad de Residencia'])):
        ax.text(value-dx, i,     name,           size=14, weight=600, ha='right', va='bottom')
        ax.text(value-dx, i-.25, name, size=10, color='#444444', ha='right', va='baseline')
        ax.text(value+dx, i,     f'{value:,.0f}',  size=14, ha='left',  va='center')

    ax.text(1, 0.4, año, transform=ax.transAxes, color='#777777', size=46, ha='right', weight=800)
    ax.text(0, 1.06, 'Número de defunciones', transform=ax.transAxes, size=12, color='#777777')
    ax.xaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))
    ax.xaxis.set_ticks_position('top')
    ax.tick_params(axis='x', colors='#777777', labelsize=12)
    ax.set_yticks([])
    ax.margins(0, 0.01)
    ax.grid(which='major', axis='x', linestyle='-')
    ax.set_axisbelow(True)
    ax.text(0, 1.12, 'Defunciones por enfermedades cardiovasculares',
            transform=ax.transAxes, size=24, weight=600, ha='left')
    ax.text(1, 0, 'Fuente: INEGI', transform=ax.transAxes, ha='right',
            color='#777777', bbox=dict(facecolor='white', alpha=0.8, edgecolor='white'))
    plt.box(False)
    
draw_barchart(2000)
```

```python
fig, ax = plt.subplots(figsize=(15, 8))

animator = animation.FuncAnimation(fig, draw_barchart, frames=range(1995, 2014))

HTML(animator.to_jshtml()) 
```
